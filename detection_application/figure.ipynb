```python
import matplotlib.pyplot as plt
import pandas as pd

# Configure LaTeX-like font settings
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    "font.size": 14
})

# Load data from CSV files
model_a_path = '/home/tim/Documents/06_Projekt_RainyNuScenes/rainynuscenes/detection_application/results/rainy_nuscenes_metrics.csv'
model_b_path = '/home/tim/Documents/06_Projekt_RainyNuScenes/rainynuscenes/detection_application/results/woodscape_metrics.csv'
model_c_path = '/home/tim/Documents/06_Projekt_RainyNuScenes/rainynuscenes/detection_application/results/both_metrics.csv'

data_a = pd.read_csv(model_a_path)
data_b = pd.read_csv(model_b_path)
data_c = pd.read_csv(model_c_path)

# Create a figure for the metrics plot
fig, ax = plt.subplots(figsize=(10, 6))

# Twin axis for IoU and Dice coefficients
ax2 = ax.twinx()

# Extract data for plotting
epochs_a = data_a['Epoch']
epochs_b = data_b['Epoch']
epochs_c = data_c['Epoch']

# Train and validation losses
ax.plot(epochs_a, data_a['Train_Loss'], label="Model A Train Loss", color="#1f77b4", linestyle='-', marker='o')
ax.plot(epochs_a, data_a['Val_Loss'], label="Model A Val Loss", color="#1f77b4", linestyle='--', marker='s')
ax.plot(epochs_b, data_b['Train_Loss'], label="Model B Train Loss", color="#ff7f0e", linestyle='-', marker='o')
ax.plot(epochs_b, data_b['Val_Loss'], label="Model B Val Loss", color="#ff7f0e", linestyle='--', marker='s')
ax.plot(epochs_c, data_c['Train_Loss'], label="Model C Train Loss", color="#00876C", linestyle='-', marker='o')
ax.plot(epochs_c, data_c['Val_Loss'], label="Model C Val Loss", color="#00876C", linestyle='--', marker='s')

ax.set_ylabel(r"\textbf{Loss}", fontsize=14)
ax.set_xlabel(r"\textbf{Epochs}", fontsize=14)

# Train and validation IoU and Dice coefficients
ax2.plot(epochs_a, data_a['Train_IoU'], label="Model A Train IoU", color="#1f77b4", linestyle='-', marker='x')
ax2.plot(epochs_a, data_a['Val_IoU'], label="Model A Val IoU", color="#1f77b4", linestyle='--', marker='d')
ax2.plot(epochs_b, data_b['Train_IoU'], label="Model B Train IoU", color="#ff7f0e", linestyle='-', marker='x')
ax2.plot(epochs_b, data_b['Val_IoU'], label="Model B Val IoU", color="#ff7f0e", linestyle='--', marker='d')
ax2.plot(epochs_c, data_c['Train_IoU'], label="Model C Train IoU", color="#00876C", linestyle='-', marker='x')
ax2.plot(epochs_c, data_c['Val_IoU'], label="Model C Val IoU", color="#00876C", linestyle='--', marker='d')

ax2.set_ylabel(r"\textbf{IoU and Dice Coefficients}", fontsize=14)

# Combine legends from both axes
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines + lines2, labels + labels2, loc='center right', fontsize=8)

# Grid and layout adjustments
ax.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Save the figure
plt.savefig('/home/tim/Documents/06_Projekt_RainyNuScenes/rainynuscenes/detection_application/results/model_comparison_metrics.pdf', format='pdf')
print("Plot saved as model_comparison_metrics.pdf")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Configure LaTeX-like font settings
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"],
    "font.size": 14
})

# Load data from CSV files
model_a_path = '/home/tim/Documents/06_Projekt_RainyNuScenes/rainynuscenes/detection_application/results/rainy_nuscenes_metrics.csv'
model_b_path = '/home/tim/Documents/06_Projekt_RainyNuScenes/rainynuscenes/detection_application/results/woodscape_metrics.csv'
model_c_path = '/home/tim/Documents/06_Projekt_RainyNuScenes/rainynuscenes/detection_application/results/both_metrics.csv'

data_a = pd.read_csv(model_a_path)
data_b = pd.read_csv(model_b_path)
data_c = pd.read_csv(model_c_path)

# Create a figure with three subplots for each model
fig, axs = plt.subplots(1, 3, figsize=(20, 6), sharey=False)

# Plot metrics for Model A
ax2_a = axs[0].twinx()
axs[0].set_title(r"\textbf{Model (A) - RainyNuScenes}", fontsize=16)
axs[0].plot(data_a['Epoch'], data_a['Train_Loss'], label="Train Loss", color="#1f77b4", linestyle='-', marker='o')
axs[0].plot(data_a['Epoch'], data_a['Val_Loss'], label="Val Loss", color="#1f77b4", linestyle='--', marker='s')
ax2_a.plot(data_a['Epoch'], data_a['Train_IoU'], label="Train IoU", color="#ff7f0e", linestyle='-', marker='x')
ax2_a.plot(data_a['Epoch'], data_a['Val_IoU'], label="Val IoU", color="#ff7f0e", linestyle='--', marker='d')
ax2_a.plot(data_a['Epoch'], data_a['Train_Dice'], label="Train Dice", color="#2ca02c", linestyle='-', marker='^')
ax2_a.plot(data_a['Epoch'], data_a['Val_Dice'], label="Val Dice", color="#2ca02c", linestyle='--', marker='v')
axs[0].set_xlabel(r"\textbf{Epochs}", fontsize=12)
axs[0].set_ylabel(r"\textbf{Loss}", fontsize=12)
#ax2_a.set_ylabel(r"\textbf{IoU and Dice}", fontsize=12)
axs[0].grid(True, linestyle='--', alpha=0.6)

# Plot metrics for Model B
ax2_b = axs[1].twinx()
axs[1].set_title(r"\textbf{Model (B) - WoodScape}", fontsize=16)
axs[1].plot(data_b['Epoch'], data_b['Train_Loss'], label="Train Loss", color="#1f77b4", linestyle='-', marker='o')
axs[1].plot(data_b['Epoch'], data_b['Val_Loss'], label="Val Loss", color="#1f77b4", linestyle='--', marker='s')
ax2_b.plot(data_b['Epoch'], data_b['Train_IoU'], label="Train IoU", color="#ff7f0e", linestyle='-', marker='x')
ax2_b.plot(data_b['Epoch'], data_b['Val_IoU'], label="Val IoU", color="#ff7f0e", linestyle='--', marker='d')
ax2_b.plot(data_b['Epoch'], data_b['Train_Dice'], label="Train Dice", color="#2ca02c", linestyle='-', marker='^')
ax2_b.plot(data_b['Epoch'], data_b['Val_Dice'], label="Val Dice", color="#2ca02c", linestyle='--', marker='v')
axs[1].set_xlabel(r"\textbf{Epochs}", fontsize=12)
axs[1].grid(True, linestyle='--', alpha=0.6)

# Plot metrics for Model C
ax2_c = axs[2].twinx()
axs[2].set_title(r"\textbf{Model (C) - RainyNuScenes+WoodScape}", fontsize=16)
axs[2].plot(data_c['Epoch'], data_c['Train_Loss'], label="Train Loss", color="#1f77b4", linestyle='-', marker='o')
axs[2].plot(data_c['Epoch'], data_c['Val_Loss'], label="Val Loss", color="#1f77b4", linestyle='--', marker='s')
ax2_c.plot(data_c['Epoch'], data_c['Train_IoU'], label="Train IoU", color="#ff7f0e", linestyle='-', marker='x')
ax2_c.plot(data_c['Epoch'], data_c['Val_IoU'], label="Val IoU", color="#ff7f0e", linestyle='--', marker='d')
ax2_c.plot(data_c['Epoch'], data_c['Train_Dice'], label="Train Dice", color="#2ca02c", linestyle='-', marker='^')
ax2_c.plot(data_c['Epoch'], data_c['Val_Dice'], label="Val Dice", color="#2ca02c", linestyle='--', marker='v')
axs[2].set_xlabel(r"\textbf{Epochs}", fontsize=12)
ax2_c.set_ylabel(r"\textbf{IoU and Dice}", fontsize=12)
axs[2].grid(True, linestyle='--', alpha=0.6)


# Add the legend inside the last plot (Model C)
lines_c, labels_c = axs[2].get_legend_handles_labels()
lines2_c, labels2_c = ax2_c.get_legend_handles_labels()
axs[2].legend(lines_c + lines2_c, labels_c + labels2_c, loc='center right', fontsize=10)




# Adjust layout and save the figure
plt.tight_layout()
plt.savefig('/home/tim/Documents/06_Projekt_RainyNuScenes/rainynuscenes/detection_application/results/model_comparison.pdf', format='pdf')
print("Plot saved as model_comparison_split_with_dice_corrected_legend_middle_right.pdf")
```

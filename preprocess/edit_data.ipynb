```python
import os

def remove_last_png(filename):
    if filename.endswith('.png'):
        # Remove the last '.png'
        new_filename = filename[:-4]  # Remove the last 4 characters
        return new_filename
    return filename  # Return the original filename if it doesn't end with .png

def process_directory(path):
    for root, dirs, files in os.walk(path):
        for file in files:
            if file.endswith('.png'):
                old_path = os.path.join(root, file)
                new_filename = remove_last_png(file)
                new_path = os.path.join(root, new_filename)
                os.rename(old_path, new_path)
                print(f"Renamed: {old_path} -> {new_path}")
```

## Validation WoodScape

```python
import os
import random
import shutil

# Paths
image_path = "/home/tim/Documents/06_Projekt_RainyNuScenes/rainynuscenes/data/soiling_woodscape_data/train/droplet_rgb"  # Replace with the path to your images
mask_path = "/home/tim/Documents/06_Projekt_RainyNuScenes/rainynuscenes/data/soiling_woodscape_data/train/droplet_masks"  # Replace with the path to your masks
val_image_path = "/home/tim/Documents/06_Projekt_RainyNuScenes/rainynuscenes/data/soiling_woodscape_data/val/droplet_rgb"  # Replace with the desired validation images path
val_label_path = "/home/tim/Documents/06_Projekt_RainyNuScenes/rainynuscenes/data/soiling_woodscape_data/val/droplet_masks"  # Replace with the desired validation masks path

# Create directories if they don't exist
os.makedirs(val_image_path, exist_ok=True)
os.makedirs(val_label_path, exist_ok=True)

# Get all image file names (assuming the masks have the same names)
image_files = [f for f in os.listdir(image_path) if os.path.isfile(os.path.join(image_path, f))]

# Calculate 20% of the dataset for validation
val_size = int(0.2 * len(image_files))

# Randomly select 20% of the images
val_images = random.sample(image_files, val_size)

# Move selected images and their corresponding masks to validation directories
for image in val_images:
    # Source paths
    image_src = os.path.join(image_path, image)
    mask_src = os.path.join(mask_path, image)  # Assuming masks have the same name

    # Destination paths
    image_dest = os.path.join(val_image_path, image)
    mask_dest = os.path.join(val_label_path, image)

    # Move files
    shutil.move(image_src, image_dest)
    shutil.move(mask_src, mask_dest)

print(f"Moved {val_size} images and their masks to the validation partition.")
```
